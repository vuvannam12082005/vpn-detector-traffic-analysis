- Vẽ biểu đồ phân phối cho mỗi cột để so sánh giữa VPN và No-VPN

##### - Phân tích biểu đồ phân phối:

    + No-VPN: Các cột có phân phối nhiều hơn trong No-VPN có thể phản ánh đặc điểm của lưu lượng internet thông thường mà không qua VPN.

<br>

- Thực hiện kiểm định thống kê cho mỗi cột để xác định sự khác biệt có ý nghĩa.

```python
from scipy.stats import ttest_ind
from IPython.display import display

# Giả sử vpn15s và novpn15s là DataFrames của bạn chứa dữ liệu VPN và no-VPN

# Kiểm định t-test cho mỗi cột
results = []
for column in vpn15s.columns:
    if column != 'class1':  # Loại trừ cột 'class1'
        t_stat, p_val = ttest_ind(vpn15s[column].dropna(), novpn15s[column].dropna())
        results.append((column, t_stat, p_val))

# Chuyển kết quả thành DataFrame để dễ dàng xem và phân tích
results_df = pd.DataFrame(results, columns=['Column', 'T-statistic', 'P-value'])

# Hiển thị DataFrame
print(results_df)
```

##### - Phân tích kết quả kiểm định thống kê:
- Giá trị `T-statistic`: Giá trị cao (về âm hay dương) cho thấy sự khác biệt lớn giữa 2 nhóm.
- Giá trị `P-value`: Một P-value thấp (thường dưới 0.05) cho thấy sự khác biệt có ý nghĩa thống kê. Trong trường hợp trên, các cột như `duration`, `total_fiat`, `min_fiat`, v.v., có P-value rất thấp, cho thấy sự khác biệt đáng kể giữa VPN và No-VPN.
    + VD: `duration` có `T-statistic` là -10.306134 và `P-value` là 7.700747e−25 có nghĩa rằng có sự khác biệt rất lớn về thời gian kết nối (duration) giữa VPN và No-VPN

## Phase 3: Huấn luyện Model


#### Ở đây, với 2 dataset đã được xử lý, mình sẽ thử train thuật toán K-Nearest Neighbors và Decision Tree để xác định model nào hiệu quả hơn trong việc phân loại

##### K-Nearest Neighbors:


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

#read data
yes_df = pd.read_csv('clean_data/vpn15s_cleaned.csv')
no_df = pd.read_csv('clean_data/novpn15s_cleaned.csv')


# Create labels for dataset
yes_df['label'] = 1  
no_df['label'] = 0   

#combine 2 dataset
combined_df = pd.concat([yes_df, no_df], ignore_index=True)

# separate features and specify target variable
X = combined_df.drop(columns=['label'])
y = combined_df['label']

#split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#specifying the category and numerical columns
cate_cols=[cname for cname in X_train.columns if X_train[cname].dtype=="object"]
nume_cols=[cname for cname in X_train.columns if X_train[cname].dtype in ['int64','float64']]

# preprocessing to deal with different type of datas
# Preprocessing for numerical data
numerical_transformer = SimpleImputer(strategy='constant')

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, nume_cols),
        ('cat', categorical_transformer, cate_cols)
    ])


# train the model
knn = Pipeline(steps=[('preprocessor', preprocessor),
                    ('model', KNeighborsClassifier(n_neighbors=5))
                             ])
knn.fit(X_train, y_train)

# predicting some value
y_pred = knn.predict(X_test)

# evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# print the report
print(classification_report(y_test, y_pred))

# view some first prediction
print("First few predictions:", y_pred[:5])
print("First few actual values:", y_test[:5].values)

# Generate matrix
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No VPN', 'VPN'], yticklabels=['No VPN', 'VPN'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

#read data
yes_df = pd.read_csv('clean_data/vpn15s_cleaned.csv')
no_df = pd.read_csv('clean_data/novpn15s_cleaned.csv')


# Create labels for dataset
yes_df['label'] = 1  
no_df['label'] = 0   

#combine 2 dataset
combined_df = pd.concat([yes_df, no_df], ignore_index=True)

# separate features and specify target variable
X = combined_df.drop(columns=['label'])
y = combined_df['label']

#split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#specifying the category and numerical columns
cate_cols=[cname for cname in X_train.columns if X_train[cname].dtype=="object"]
nume_cols=[cname for cname in X_train.columns if X_train[cname].dtype in ['int64','float64']]

# preprocessing to deal with different type of datas
# Preprocessing for numerical data
numerical_transformer = SimpleImputer(strategy='constant')

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, nume_cols),
        ('cat', categorical_transformer, cate_cols)
    ])
cpu_cores=8
# Train the model based on best optimized args
rf=Pipeline(steps=[('preprocessor', preprocessor),
                    ('model', RandomForestClassifier(n_jobs=cpu_cores,n_estimators=7000))
                             ])
rf.fit(X_train,y_train)

# predicting some value
y_pred = rf.predict(X_test)

# evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# print the report
print(classification_report(y_test, y_pred))

# view some first prediction
print("First few predictions:", y_pred[:5])
print("First few actual values:", y_test[:5].values)

# Generate matrix
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No VPN', 'VPN'], yticklabels=['No VPN', 'VPN'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
